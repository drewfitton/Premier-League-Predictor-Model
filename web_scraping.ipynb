```python
import requests
import os
import shutil
```

```python
url_start = "https://fbref.com/en/comps/9/Premier-League-Stats.html"

data = requests.get(url_start)
with open("Teamstats.html", "w+") as f:
    f.write(data.text)
```

```python
from bs4 import BeautifulSoup
```

```python
with open("Teamstats.html") as f:
    page = f.read()
    
```

```python
soup = BeautifulSoup(page, "html.parser")
league_table = soup.find_all(id="results111601_overall")[0]
```

```python
league_table
```

```python
import pandas as pd
```

```python
leaguetable = pd.read_html(str(league_table))
```

```python
leaguetable
```

```python
league = pd.concat(leaguetable)
```

```python
league.to_csv("league_table.csv")
```

```python
url = "https://fbref.com/en/comps/9/schedule/Premier-League-Scores-and-Fixtures.html"

newdata = requests.get(url)
with open("fixtures.html", "w+") as f:
    f.write(newdata.text)
    
```

```python
with open("fixtures.html") as f:
    page = f.read()

soup = BeautifulSoup(page, "html.parser")
matches = soup.find_all(id="sched_11160_1")[0]
matchtable = pd.read_html(str(matches))

fixtures = pd.concat(matchtable)
fixtures.to_csv("match_table.csv")
```

```python
future_url = "https://fbref.com/en/comps/9/11566/schedule/2022-2023-Premier-League-Scores-and-Fixtures"

newfix = requests.get(future_url)
with open("future_fixtures.html", "w+") as f:
    f.write(newfix.text)

with open("future_fixtures.html") as f:
    page = f.read()
    
soup = BeautifulSoup(page, "html.parser")
matches = soup.find_all(id="sched_11566_1") [0]
matchtable = pd.read_html(str(matches))

futurefix = pd.concat(matchtable)
futurefix.to_csv("future_matches_table.csv")
```
