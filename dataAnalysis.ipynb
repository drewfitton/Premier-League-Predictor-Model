```python
import pandas as pd
from scipy.stats import poisson

table = pd.read_csv("combined.csv")
```

```python
table
```

```python
futureTable = pd.read_csv("future_matches_table.csv")

for a, b in futureTable.iterrows():
    
    home = b["Home"]
    away = b["Away"]
    
    if not home in table:
        continue
    elif not away in table:
        continue

    for i, j in table.iterrows():
        if (j["Squad"] == home):
            totalHGF = j["GF"]
            totalHGA = j["GA"]
            totalHMP = j["MP"]
            totalHGP90 = totalHGF / totalHMP
            totalHConcededP90 = totalHGA / totalHMP
            homeGF = j["HomeGF"]
            homeMP = j["HomeMP"]
            homeGA = j["HomeGA"]
            homeRK = j["Rk"]
            homeGP90 = homeGF / homeMP
            homeConcededP90 = homeGA / homeMP
            homeGP90 = (homeGP90 * .6) + (totalHGP90 * .4)
            homeConcededP90 = (homeConcededP90 * .6) + (totalHConcededP90 * .4)
            lastScore = j[away]
        if (j["Squad"] == away):
            totalAGF = j["GF"]
            totalAGA = j["GA"]
            totalAMP = j["MP"]
            totalAGP90 = totalAGF / totalAMP
            totalAConcededP90 = totalAGA / totalAMP
            awayGF = j["AwayGF"]
            awayMP = j["Away"]
            awayGA = j["AwayGA"]
            awayRK = j["Rk"]
            awayGP90 = awayGF / awayMP
            awayConcededP90 = awayGA / awayMP
            awayGP90 = (awayGP90 * .6) + (totalAGP90 * .4)
            awayConcededP90 = (awayConcededP90 * .6) + (totalAConcededP90 * .4)
            lastScoreOpp = j[home]
        
    odds = 0
    expected = 0
    for i in range(10):
        hgfOdds = poisson.pmf(i, homeGP90)
        agaOdds = poisson.pmf(i, awayConcededP90)
        if (hgfOdds + agaOdds > odds):
            odds = hgfOdds + agaOdds
            expected = i

    awayOdds = 0
    awayExpected = 0
    for i in range(10):
        agfOdds = poisson.pmf(i, awayGP90)
        hgaOdds = poisson.pmf(i, homeConcededP90)
        if (agfOdds + hgaOdds > awayOdds):
            awayOdds = agfOdds + hgaOdds
            awayExpected = i

    if homeRK - awayRK >= 14:
        awayExpected += 2
    elif homeRK - awayRK >= 7:
        awayExpected += 1
    elif awayRK - homeRK >= 14:
        expected += 2
    elif awayRK - homeRK >= 7:
        expected += 1

    if not lastScore == "0":
        x = lastScore.split("–")
        homeG = int(x[0])
        awayG = int(x[1])
        if not lastScoreOpp == "0":
            y = lastScoreOpp.split("–")
            homeG2 = int(y[1])
            awayG2 = int(y[0])
            homeG = int((homeG + homeG2) / 2)
            awayG = int((awayG + awayG2) / 2)

    if (homeG - expected) >= 3:
        expected += 2
    elif (homeG - expected) >= 1:
        expected += 1
    elif (awayG - awayExpected) >= 3:
        awayExpected += 2
    elif (awayG - awayExpected) >= 1:
        awayExpected += 1
    
    futureTable.at[a, "xG"] = str(expected)
    futureTable.at[a, "xG.1"] = str(awayExpected)
```
