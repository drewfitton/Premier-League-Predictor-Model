```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
fixtures = pd.read_csv("2022-2023_Fixtures.csv")
```

```python
last_table = pd.read_csv("league_table.csv")
new_table = pd.DataFrame(columns= ["Rk", "Squad", "MP", "W", "D", "L", "GF", "GA", "GD", "Pts", "Pts/MP"])
```

```python
new_table
```

```python

for x, y in last_table.iterrows():
    if x == 17:
        team = "Nott'ham Forest"
    elif x == 18:
        team = "Fulham"
    elif x == 19:
        team = "Bournemouth"
    else:
        team = y["Squad"]
    wins = 0
    draws = 0
    losses = 0
    mp = 0
    gf = 0
    ga = 0
    gd = 0
    pts = 0
    ptsmp = 0
    
    for i, j in fixtures.iterrows():
        if j["Home"] == team:
            diff = j["xG"] - j["xG.1"]
            gf += j["xG"]
            mp += 1
            ga += j["xG.1"]
            gd += diff
            if diff > 0:
                wins += 1
                pts += 3
            elif diff < 0:
                losses += 1
            else:
                draws += 1
                pts += 1
        elif j["Away"] == team:
            diff = j["xG.1"] - j["xG"]
            gf += j["xG.1"]
            mp += 1
            ga += j["xG"]
            gd += diff
            if diff > 0:
                wins += 1
                pts += 3
            elif diff < 0:
                losses += 1
            else:
                draws += 1
                pts += 1
    ptsmp = pts / mp
    new_table.loc[len(new_table.index)] = [0, team, mp, wins, draws, losses, gf, ga, gd, pts, ptsmp]
```

```python
new_table
```

```python
new_table = new_table.sort_values(by = 'GD', ascending = 1)
new_table = new_table.sort_values(by = 'Pts', ascending = 0)
new_table = new_table.reset_index()
```

```python
del new_table["index"]
```

```python
new_table
```

```python
for i, j in new_table.iterrows():
    new_table.at[i, "Rk"] = i + 1
    new_table.at[i, "Pts/MP"] = round(j["Pts/MP"], 2)
    new_table.at[i, "GF"] = int(j["GF"])
    new_table.at[i, "GA"] = int(j["GA"])
    new_table.at[i, "GD"] = int(j["GD"])
```

```python
new_table.to_csv("2022_2023_table.csv")
```
